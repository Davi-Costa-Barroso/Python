```python
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np
# Taxa de desemprego de 2019 a 2021
# Assunto: Distribuição Normal (Gaussiana)

# Pega os dados da planilha
dados = pd.read_csv("desemprego.csv")
y = dados.valor
print(y)
```

```python
# Calcula a media(μ)
media = sum(y)/len(y)

# Calcula o desvio padrão(σ)
sigma = y.std()

# Maior e menor valor
v_max = max(y)
v_min = min(y)

print("Desvio padrão(σ):", round(sigma, 3))
print("Media(μ):", media)
```

```python
# Intervalos para visualizar melhor a curva em formato de sino
x = np.linspace(v_min-2, v_max+2)
# Formula da curva de gauss
formula = 1/(sigma*(2*math.pi)**0.5)*np.exp(-0.5*((x-media)/sigma)**2)
# Determina onde a formula será colocada no eixo x
x = np.linspace(v_min, v_max)

# Plota o grafico
# figsize=(largura, altura)
plt.figure(figsize=(12,6))
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] = 18
plt.title("Desemprego 2019 - 2021 (Brasil)", fontsize=22)
plt.xlabel("Taxa de desemprego (%)", fontsize=20)
plt.ylabel("f(x)", fontsize=20)
plt.plot(x, formula, color="Black")
# Preenche a area do desvio padrão(σ)
plt.fill_between(x, formula, alpha=0.3, where=(x >= media), label="Desvio padrão=1.357")
plt.legend()
plt.show()
```
