```python
import matplotlib.pyplot as plt 

# Taxa de desemprego por região
x = ["Norte", "Nordeste", "Centro-Oeste", "Sudeste", "Sul"]
y = [11.2, 14.7, 8.4, 11.2, 6.7]
```

```python
#Grafico de barras
plt.bar(x,y)
plt.title("Taxa de Desocupação nas Grandes Regiões do Brasil, 4º trimestre 2021")
plt.ylabel("Porcentagem (%)")
plt.show()
```
