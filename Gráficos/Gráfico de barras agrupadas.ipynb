```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

# Dados de desemprego por sexo dos últimos trimestres de 2018 a 2021
dados = pd.read_excel("desemprego_por_sexo.xlsx")
mulheres = dados["Mulheres (%)"]
homens = dados["Homens (%)"]
```

```python
#Largura das barras
barWidth= 0.4

#Tamanho do grafico
plt.figure(figsize=(12, 4))

# Cria uma lista de 0 até a quantidade de valores que tem em homens
r1 = np.arange(len(homens)) 
r2 = [x + barWidth for x in r1] 

#Grafico de barras agrupadas
plt.bar(r1, homens, color='blue', width=barWidth, label='Homens')
plt.bar(r2, mulheres, color='red', width=barWidth, label='Mulheres')
plt.xticks([r + barWidth for r in range(len(homens))], ['4º trim 2021', '4º trim 2020', '4º trim 2019', '4º trim 2018'])
plt.title("Distribuição percentual de pessoas desocupadas por sexo nos últimos 4 anos")
plt.ylabel("Porcentagem (%)")
plt.legend()
plt.show()
```
