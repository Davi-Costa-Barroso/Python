```python
import matplotlib.pyplot as plt 
import pandas as pd

# Variação trimestral de desemprego de 2019 a 2021
dados = pd.read_csv("desemprego-estatistica-trimestral.csv")
x = dados.periodo
y = dados.valor
```

```python
# Determina o tamanho do gráfico
plt.figure(figsize=(12,4))

# Plota o gráfico de linhas
plt.plot(x, y, color='red') 
plt.scatter(x, y, color='red')
plt.title("Variação trimestral de desemprego - Brasil") 
plt.ylabel("Porcentagem (%)")
plt.grid()
plt.show()
```
