```python
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np

# Renda per Capita de 2008 a 2019
# Assunto: Distribuição Normal ou curva Gaussiana

# Pega os dados da planilha
dados = pd.read_csv("pib-per-capita-r.csv")
y = dados.valor
print(y)
```

```python
# Calcula a media(μ)
media = sum(y)/len(y)
# Calcula o desvio padrão(σ)
sigma = y.std()
# Separa o maior e menor valor
v_max = max(y)
v_min = min(y)

print("Valor minimo: ", v_min)
print("Valor maximo: ", v_max)
print("Desvio padrão(σ):", round(sigma, 3))
print("Media(μ):", round(media, 3))
```

```python
# Determina os intervalos para visualizar melhor a curva em formato de sino
x = np.linspace(v_min-5000, v_max+5000)

# Formula da curva de gauss
formula = 1/(sigma*(2*math.pi)**0.5)*np.exp(-0.5*((x-media)/sigma)**2)
# Determina onde a formula será plotada no eixo x
x = np.linspace(v_min, v_max)

# Plota o grafico
# figsize=(largura, altura)
plt.figure(figsize=(12,6))
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] = 18
plt.title("PIB per capita 2008 - 2019 (Brasil)", fontsize=22)
plt.xlabel("Valor (R$)", fontsize=20)
plt.ylabel("f(x)", fontsize=20)
plt.plot(x, formula, color="Black")

# Preenche a area do desvio padrão(σ)
plt.fill_between(x, formula, alpha=0.3, where=(x >= media), label="Desvio padrão", color="blue")
plt.legend()
plt.show()
```
