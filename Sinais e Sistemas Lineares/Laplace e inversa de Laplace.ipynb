```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
t, s = sp.symbols('t, s')
```

```python
f = sp.exp(3*t)
f
```

```python
#Transforma f(t) para o domínio de Laplace F(S)
F = sp.laplace_transform(f, t, s, noconds=True)
#Imprime o resultado no domínio S
F
```

```python
# Utilize para dividir em frações parciais
#F.apart(s)
```

```python
# Transformada Inversa de Laplace
f1 = sp.inverse_laplace_transform(F, s, t)
f1
```
